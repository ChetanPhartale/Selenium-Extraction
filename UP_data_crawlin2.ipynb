```python
from bs4 import BeautifulSoup
from seleniumwire import webdriver
import pandas as pd
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from time import sleep
import warnings
import os
from selenium.webdriver.support.ui import Select

warnings.filterwarnings('ignore')

import io

chrome_options = webdriver.ChromeOptions()
# chrome_options.add_experimental_option('prefs', prefs)
chrome_options.add_argument('--kiosk-printing')
# chrome_options.add_argument('--headless')
chrome_options.add_argument('log-level=3')
# chrome_options.add_argument('--disable-gpu')
chrome_options.add_argument("--window-size=1920,1080")

import easyocr

from PIL import Image 
from selenium.webdriver.chrome.service import Service
import numpy as np
CHROMEDRIVER_PATH =r"C:\Users\SY-DKT-MUM-112\Chromewebdriver\chromedriver-win64"
service = Service(executable_path = CHROMEDRIVER_PATH)
driver = webdriver.Chrome(chrome_options=chrome_options)
def cap_slover():
    img2 = driver.find_elements_by_tag_name('img')[2]

    captcha2= img2.screenshot_as_png
    reader = easyocr.Reader(['en'], gpu=False)

    im2 = Image.open(io.BytesIO((captcha2)))
    im2 = np.asarray(im2)
    captcha= (reader.readtext(im2))[0][1]
    return captcha
def cap_slover_main():
    #img2 = driver.find_element_by_id("captchaImage")
    img2 = driver.find_element("id","captchaImage")

    captcha2= img2.screenshot_as_png
    
    reader = easyocr.Reader(['en'], gpu=False)
    im2 = Image.open(io.BytesIO((captcha2)))
    im2 = np.asarray(im2)
    captcha= (reader.readtext(im2))[0][1]
    return captcha
```

```python
from selenium.webdriver.support.select import Select
district = []

select = Select(driver.find_element("id",'districtCode'))

for opt in select.options:
    district.append([opt.text,opt.get_attribute("value")])

print(district[0])




```

```python
district_id = []
for i in district:
    district_id.append(i[1])
print(district_id)

district_name = []
for i in district:
    district_name.append(i[0])
print(district_name)
```

```python

tehsil_1 = []
village_1 = []


driver.implicitly_wait(10)
wait = WebDriverWait(driver, 10)

driver.find_element("id", 'districtCode').click()


for dist in district_name[:5]:
    select = Select(driver.find_element("id", 'districtCode'))
    select.select_by_visible_text(dist)

    # Wait for the tehsil dropdown list to be visible
    wait.until(EC.visibility_of_element_located((By.ID, 'tehsilCode3')))
    tehsil = []
    select = Select(driver.find_element("id", 'tehsilCode3'))
    for opt in select.options:
        tehsil.append([opt.text, opt.get_attribute("value")])
    #print(tehsil)

    #tehsil_1.append([tehsil,dist])


    tehsil_id = []
    for tehl in tehsil:
        tehsil_id.append(tehl[1])

    tehsil_name = []
    for tehl in tehsil:
        tehsil_name.append(tehl[0])

    

    for teh in tehsil_name:
    #     #print(teh)
        
    #     driver.find_element("id", 'tehsilCode3')
        select = Select(driver.find_element("id", 'tehsilCode3'))
        select.select_by_visible_text(teh)
        wait.until(EC.visibility_of_element_located((By.ID, 'villageCode3')))

        village = []

        select = Select(driver.find_element("id", 'villageCode3'))
        for opt in select.options:
            village.append([opt.text, opt.get_attribute("value")])

        # village_id = []
        # for vil in village:
        #     village_id.append(village[1])

        # village_name = []
        # for vil in village:
        #     village_name.append(village[0])

    #     print(village)
        village_1.append(village)
        
    tehsil_1.append([tehsil,[village_1]])

```
