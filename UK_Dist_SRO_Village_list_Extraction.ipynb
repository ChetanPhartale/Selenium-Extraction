```python

options = webdriver.ChromeOptions() 
options.headless = False
driver = webdriver.Chrome(options=options)
wait = WebDriverWait(driver, 10)

driver.get("https://online.eregistrationukgov.in/e_search/Frm_vill_Wise.aspx#")

from selenium.webdriver.support.select import Select
District = []

driver.find_element(By.XPATH,'//*[@id="MainContent_ddl_dis"]').click()
select = Select(driver.find_element(By.XPATH,'//*[@id="MainContent_ddl_dis"]'))
for opt in select.options:
    District.append([opt.text,opt.get_attribute("value")])

District.pop(0)    

District_id = []
for i in District:
    District_id.append(i[1])

District_name = []
for i in District:
    District_name.append(i[0])

SRO = []
Village1 = []
Final_list = []

for dist in District[3:13]:
    driver.find_element(By.XPATH,'//*[@id="MainContent_ddl_dis"]').click()
    select = Select(driver.find_element(By.XPATH,'//*[@id="MainContent_ddl_dis"]'))
    select.select_by_visible_text(dist[0])
    b = dist[0]
    a = dist[1]
    
    sleep (2)
    
    SRO = []
    WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, "//*[@id='MainContent_ddl_sro']")))
    driver.implicitly_wait(2)
    driver.find_element(By.XPATH,'//*[@id="MainContent_ddl_sro"]').click()
    select = Select(driver.find_element(By.XPATH,'//*[@id="MainContent_ddl_sro"]'))
    for opt in select.options:
        SRO.append([opt.text,opt.get_attribute("value")])
    SRO.pop(0)
    
    SRO_id = []
    for i in SRO:
        SRO_id.append(i[1])

    SRO_name = []
    for i in SRO:
        SRO_name.append(i[0])
        
    for sr in SRO:
        driver.find_element(By.XPATH,'//*[@id="MainContent_ddl_sro"]').click()
        select = Select(driver.find_element(By.XPATH,'//*[@id="MainContent_ddl_sro"]'))
        select.select_by_visible_text(sr[0])
        d = sr[0]
        c = sr[1]
        
        sleep(2)
        
        year = []
        
        driver.find_element(By.XPATH,'//*[@id="MainContent_dd_regyear"]').click()
        select = Select(driver.find_element(By.XPATH,'//*[@id="MainContent_dd_regyear"]'))
        for opt in select.options:
            year.append([opt.text,opt.get_attribute("value")])
        year.pop(0)
#         print(year)
        select.select_by_visible_text(year[-2][1])
        
        sleep(2)
        
        Village = []
        
        WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, "//*[@id='MainContent_dd_vill']")))
        driver.implicitly_wait(2)
        driver.find_element(By.XPATH,'//*[@id="MainContent_dd_vill"]').click()
        select = Select(driver.find_element(By.XPATH,'//*[@id="MainContent_dd_vill"]'))
        for opt in select.options:
            Village.append([opt.text,opt.get_attribute("value")])
        Village.pop(0)

        Village_id = []
        for i in Village:
            Village_id.append(i[1])

        Village_name = []
        for i in SRO:
            Village_name.append(i[0])
            
        for vil in Village:
            f = vil[0]
            e = vil[1]
            Final_list.append([a,b,c,d,e,f])
        
        print(a,b,c,d)
        sleep(0.5)

    
```
