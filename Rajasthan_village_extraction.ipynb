```python

options = webdriver.ChromeOptions() 
options.headless = False
driver = webdriver.Chrome(options=options)
wait = WebDriverWait(driver, 10)

driver.get("https://epanjiyan.rajasthan.gov.in/")

from selenium.webdriver.support.select import Select

Final_list = []
z = "Urban"
y = "Rural"
driver.find_element(By.XPATH,'//*[@id="ContentPlaceHolder1_lnkesearch"]').click()
sleep(2)
district = []

driver.find_element(By.XPATH,'//*[@id="ContentPlaceHolder1_rbtrural"]').click()

sleep(2)

select = Select(driver.find_element(By.XPATH,'//*[@id="ContentPlaceHolder1_ddlDistrict"]'))
for opt in select.options:
    district.append([opt.get_attribute("value"),opt.get_attribute('text')])
district.pop(0) 

for dis in district[28:50]:
#     driver.find_element(By.XPATH,'//*[@id="ContentPlaceHolder1_ddlDistrict"]').click()
    driver.find_element(By.XPATH,'//*[@id="ContentPlaceHolder1_ddlDistrict_chosen"]').click()
    driver.find_element(By.XPATH,'//*[@id="ContentPlaceHolder1_ddlDistrict_chosen"]/div/div/input').send_keys(str(dis[1]))
    driver.find_element(By.XPATH,'//*[@id="ContentPlaceHolder1_ddlDistrict_chosen"]/div/ul/li').click()
#     select = Select(driver.find_element(By.XPATH,'//*[@id="ContentPlaceHolder1_ddlDistrict_chosen"]/div/ul'))
#     select.select_by_value(str(dis[0]))
    
    a = dis[0]
    b = dis[1]
    
    sleep(2)
    
    Tehsil = []
    select = Select(driver.find_element(By.XPATH,'//*[@id="ContentPlaceHolder1_ddlTehsil"]'))   
    for opt in select.options:
        Tehsil.append([opt.get_attribute("value"),opt.get_attribute('text')])
    Tehsil.pop(0) 
    
    sleep(1)
    
    for teh in Tehsil:
        driver.find_element(By.XPATH,'//*[@id="ContentPlaceHolder1_ddlTehsil_chosen"]').click()
        driver.find_element(By.XPATH,'//*[@id="ContentPlaceHolder1_ddlTehsil_chosen"]/div/div/input').send_keys(str(teh[1]))
        driver.find_element(By.XPATH,'//*[@id="ContentPlaceHolder1_ddlTehsil_chosen"]/div/ul/li').click()
        c = teh[0]
        d = teh[1]
        
        sleep(2)
        
        SRO = []
        select = Select(driver.find_element(By.XPATH,'//*[@id="ContentPlaceHolder1_ddlSRO"]'))
        for opt in select.options:
            SRO.append([opt.get_attribute("value"),opt.get_attribute('text')])
        SRO.pop(0) 
        
        sleep(1)
        
        for sr in SRO:
            driver.find_element(By.XPATH,'//*[@id="ContentPlaceHolder1_ddlSRO_chosen"]').click()
            driver.find_element(By.XPATH,'//*[@id="ContentPlaceHolder1_ddlSRO_chosen"]/div/ul/li[2]').click()
        
            e = sr[0]
            f = sr[1]
            
            sleep(2)
            
            Zone = []
            select = Select(driver.find_element(By.XPATH,'//*[@id="ContentPlaceHolder1_ddlzonevillage"]'))
            for opt in select.options:
                Zone.append([opt.get_attribute("value"),opt.get_attribute('text')])
            Zone.pop(0) 
            
            for zone in Zone:
                g = zone[0]
                h = zone[1]
                Final_list.append([y,a,b,c,d,e,f,g,h])
                print(y,b,d,f)
            sleep(0.5)
        

```
